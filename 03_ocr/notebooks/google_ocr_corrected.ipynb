```python
# Step 2: Grammar Correction

GRAMMAR_CORRECTION_PROMPT = r"""
You are an expert Kabyle (Taqbaylit/Amazigh/Berber) language editor and proofreader.

Your task is to review and correct the following Kabyle text that was extracted via OCR.

═══════════════════════════════════════════════════════════════════
CORRECTION PRIORITIES
═══════════════════════════════════════════════════════════════════

1. FIX y vs ɣ ERRORS:
   - All 1st person singular verbs MUST end in -ɣ (not -y)
   - Examples: nniɣ, bɣiɣ, wehmeɣ, ssneɣ, fkiɣ, walaɣ, ruḥeɣ
   - Words with ɣ in root: ɣef, ɣer, ɣur, aɣyul, iɣil, taɣect
   - y- at start for 3rd person: yella, yenna, yemma

2. FIX UNDERDOTTED CONSONANTS:
   - Ensure emphatic consonants have proper underdots: ḍ, ṛ, ṣ, ṭ, ẓ, ḥ
   - Common words: aṭas, mliḥ, leḥcic, amecṭuḥ, aqerruy

3. FIX SPECIAL CHARACTERS:
   - ɛ (ayin) - not e or ε
   - č (ch sound) - not c or ch
   - ǧ (j sound) - not j or dj

4. FIX SPACING AND PUNCTUATION:
   - Correct word boundaries
   - Fix hyphenation for clitics (e.g., nniɣ-as, i t-bɣiɣ)

5. FIX COMMON OCR MISTAKES:
   - Letter substitutions
   - Missing or extra characters
   - Word segmentation errors

═══════════════════════════════════════════════════════════════════
OUTPUT RULES
═══════════════════════════════════════════════════════════════════

- Output ONLY the corrected Kabyle text
- Preserve the original formatting and line breaks
- Do NOT add translations, explanations, or comments
- Do NOT change the meaning of the text

═══════════════════════════════════════════════════════════════════
TEXT TO CORRECT:
═══════════════════════════════════════════════════════════════════

"""

print("Step 2: Correcting grammar...")

# Send the OCR text for grammar correction
correction_response = model.generate_content(
    GRAMMAR_CORRECTION_PROMPT + ocr_text
)

corrected_text = correction_response.text

# Save corrected output
with open("ocr_corrected.md", 'w', encoding='utf-8') as fil:
    fil.write(f'{corrected_text}\n')

print('Corrected output saved to ocr_corrected.md')
print("\n--- Corrected Text ---\n")
print(corrected_text)
```
